```python
import numpy as np, scipy.linalg as sl
import scipy.special as ss, math
import matplotlib.pylab as plt
import piccard as pic
import os, glob
```

```python
t2df = pic.DataFile('J0030.h5')
t2df.addTempoPulsar('./J0030+0451.par', './J0030+0451.tim')
```

```python
#%load_ext line_profiler
#%load_ext memory_profiler
```

```python
likob = pic.ptaLikelihood('J0030.h5')
modeldict = likob.makeModelDict(nfreqs=30, incRedNoise=True, noiseModel='powerlaw', varyEfac=True, separateEfacs=False, \
                                incEquad=False, separateEquads=False, \
                                ndmfreqs=0, incDM=False, dmModel='dmspectrum', \
                                incCEquad=False, separateCEquads=False, \
                                incBWM=False, incGWB=False, gwbModel='powerlaw', \
                                likfunc='mark3')
likob.initModel(modeldict, fromFile=False, verbose=False)
```

```python
pic.RunPTMCMC(likob, 20000, './chains/')
```

```python
pic.makeAllPlots('./chains/', './chains/', burnin=1000, skipTMP=True)
```
