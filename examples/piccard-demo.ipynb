## HDF5 files ##
Piccard does not work with raw par/tim files as Tempo/Tempo2 do. It requires the par/tim files to be processed beforehand. The pulsar data is stored in HDF5 files, which Piccard interacts with through the h5py package. HDF5 files can be made from par/tim files with Piccard thanks to the libstempo interface to Tempo2

### Creating the HDF5 files ###
Assume we have a pulsar (J0030 from the first IPTA Mock Data Challenge, open1, here), and read it in an HDF5 file

```python
# Creating hdf5 files goes through the DataFile class
! rm J0030.h5   # Delete it if it exists (otherwise causes an error)
t2df = pic.DataFile('J0030.h5')
t2df.addTempoPulsar('mdc1-open1/J0030+0451.par', 'mdc1-open1/J0030+0451.tim')

# This automatically writes and closes the file. 
```

### Creating multi-pulsar sets ###
A single HDF5 file can host multiple pulsars. Say we want to place all pulsars of mdc1-open1 in one HDF5 file, we would do:

```python
! rm mdc1-open1.h5   # Delete it if it exists (otherwise causes an error)
t2df = pic.DataFile('mdc1-open1.h5')
for infile in glob.glob(os.path.join('mdc1-open1/', '*.par') ):
    filename = os.path.splitext(infile)
    parfile = filename[0]+'.par'
    timfile = filename[0]+'.tim'
    
    t2df.addTempoPulsar(parfile, timfile)
```

## Creating a model for a dataset ##
The goal is to create a likelihood function. In Piccard, likelihood functions are represented by objects of the type ptaLikelihood. A likelihood function needs both observations, and a description of the model. The data is read from HDF5 files. A model can be stored in JSON files, but we can also create models on the fly. Here we'll show how to do such things.

```python
# With a datafile, we can construct a 'likelihood' object, which is the central class of the package
likob = pic.ptaLikelihood('J0030.h5')
```

```python
# Create a very simple Python dictionary that represents a model
modeldict = likob.makeModelDict(varyEfac=True, nfreqs=10, incRedNoise=True, noiseModel='powerlaw')
print modeldict    # The dictionary is human-readable
```

```python
# We now have created a model for J0030 with a varying Efac, and a red-noise (10 frequencies) modelled similarly
# to what is done in the Tempo2 'spectralModal' plugin. We can apply the model as follows
likob.initModel(modeldict)

# The parameters of this model are viewable in the pardes dictionary.
# NOTE: likob.pardes is only descriptive. It is not used to change the model
for pd in likob.pardes:
    print pd

print "Number of dimensions: ", likob.dimensions
```

#### Remember the pardes parameter descriptions. ####
Especially the 'index' field could become useful. It is the unique index of the parameter when calling the likelihood function. You can see here that the 'low-frequency-cutoff' does not have a non-negative index, since it is a fixed parameter. In the model dictionary you can see that bvary for this signal had [True, True, False], with the last parameter the 'low-frequency-cutoff'.

```python
# The model can be saved to a JSON file with
likob.writeModelToFile('J0030.json')

# A model can be read-in with
likob.initModelFromFile('J0030.json')

# Given a likelihood object, the model dictionary can be obtained with
modeldict = likob.getModelDict()

# Finally, a likelihood object can also be created with a model included
likob = pic.ptaLikelihood('J0030.h5', 'J0030.json')
```

```python
# The parameter space is 3 dimensional: (efac, red-noise-amplitude, red-noise-spectral-index)
print "Minimum bound:", likob.pmin
print "Maximum bound:", likob.pmax
print "Starting position:", likob.pstart
```

```python
# With a fully initialised likelihood object, we have access to the likelihood, prior, and posterior
print likob.logprior(likob.pstart), likob.loglikelihood(likob.pstart), likob.logposterior(likob.pstart)
print likob.logposterior([1.0, np.log10(5e-14), 4.33])   # These are the injected values for mdc1-open1
```

## Combining models from separate JSON files ##
A very likely scenario to occur is that of a staged analysis: first the pulsars will be analysed individually in some noise analysis. After the noise analysis has been properly done, the results will all be formatted in (*.par, *.tim, *.h5, *.json) files. One of those for each pulsar. However, at that point, the models should all be combined, and a single h5/JSON file will be required. The combination utilities in Piccard allow one to do just that

```python
# We will create three separate sets for the first mock data challenge: J0xxx.par, J1xxx.par, J2xxx.par.
! rm mdc1-J0xxx.h5   # Delete it if it exists (otherwise causes an error)
! rm mdc1-J1xxx.h5
! rm mdc1-J2xxx.h5

t2df = pic.DataFile('mdc1-J0xxx.h5')
for infile in glob.glob(os.path.join('mdc1-open1/', 'J0*.par') ):
    filename = os.path.splitext(infile)
    parfile = filename[0]+'.par'
    timfile = filename[0]+'.tim'
    
    t2df.addTempoPulsar(parfile, timfile)

t2df = pic.DataFile('mdc1-J1xxx.h5')
for infile in glob.glob(os.path.join('mdc1-open1/', 'J1*.par') ):
    filename = os.path.splitext(infile)
    parfile = filename[0]+'.par'
    timfile = filename[0]+'.tim'
    
    t2df.addTempoPulsar(parfile, timfile)

t2df = pic.DataFile('mdc1-J2xxx.h5')
for infile in glob.glob(os.path.join('mdc1-open1/', 'J2*.par') ):
    filename = os.path.splitext(infile)
    parfile = filename[0]+'.par'
    timfile = filename[0]+'.tim'
    
    t2df.addTempoPulsar(parfile, timfile)
```

```python
likoblist = [] ; modellist = []

# Read in the HDF5 files
likoblist.append(pic.ptaLikelihood('mdc1-J0xxx.h5'))
likoblist.append(pic.ptaLikelihood('mdc1-J1xxx.h5'))
likoblist.append(pic.ptaLikelihood('mdc1-J2xxx.h5'))


# The models are very similar
modellist.append(likoblist[0].makeModelDict(varyEfac=True, nfreqs=5, incRedNoise=True, noiseModel='powerlaw'))
modellist.append(likoblist[1].makeModelDict(varyEfac=True, nfreqs=10, incRedNoise=True, noiseModel='spectrum'))
modellist.append(likoblist[2].makeModelDict(varyEfac=True, nfreqs=20, incRedNoise=True, noiseModel='spectralModel'))

# If we would like, we could apply the models, and write to disk
# likoblist[0].initModel(modellist[0])
# likoblist[0].writeModelToFile('mdc1-J0xxx.json')
```

```python
# Ok, fine, we have the separate models. Now we can combine them like this
likob = pic.ptaLikelihood('mdc1-open1.h5') # The full set
fullmodel = likob.combineModelDicts(*modellist)
```

```python
likob.initModel(fullmodel)
likob.logposterior(likob.pstart)
```

# Simulating data #
Piccard allows the simulation of data for single pulsars and multiple pulsars through the function 'simulateFullSet'. What is required is an already existing set of par/tim files. Piccard will then take the sampling, and create new realisation of data based on some model that needs to be provided. 

```python
# We need a list of the par-files, tim-files, and the sim-files we will use.
# The par/tim files are input. the sim-files are output
parlist = [] ; timlist = [] ; simlist = []

for infile in glob.glob(os.path.join('./mdc1-open1/', '*.par') ):
    filename = os.path.splitext(infile)
    parlist.append(filename[0]+'.par')
    timlist.append(filename[0]+'.tim')
    simlist.append(filename[0]+'.sim')

# Print the first three par-files
print parlist[:3]
```

```python
# Above we called the function 'makeModelDict' using keywords given to the function.
# Here we'll use a dictionary first, since we'll need these quantities a few times
model = {'nfreqs':10, 'varyEfac':True, 'incEquad':False, 'incRedNoise':True, \
                    'noiseModel':'powerlaw', 'incGWB':True, 'gwbModel':'powerlaw'}

# Per pulsar: efac, red-noise amp, red-noise spectral index
# At the end: GWB amplitude, GWB spectral index
# The list of parameters therefore becomes:
p = len(parlist)   # The number of pulsars we read in
parameters = [1.0, -18.0, 4.33]*p + [np.log10(5e-15), 4.33]
```

```python
# Simulate the data:
! rm mockdata.h5
pic.simulateFullSet(parlist, timlist, simlist, parameters, 'mockdata.h5', **model)
```

```python
# Plot all the residuals in the file mockset.h5
likob = pic.ptaLikelihood('mockdata.h5')

plt.figure(figsize=(16,11))
for ii in range(len(likob.ptapsrs)):
    plt.subplot(9, 4,ii+1)
    plt.errorbar(likob.ptapsrs[ii].toas, likob.ptapsrs[ii].residuals, \
                 yerr=likob.ptapsrs[ii].toaerrs, fmt='.', c='b')
    
    plt.annotate(likob.ptapsrs[ii].name, \
                 xy=(0.02, 0.20), xycoords='axes fraction', \
                 bbox=dict(boxstyle="round", fc="1.0"))
    #plt.axis([53000, 55000, -5e-7, 5e-7])
    plt.grid(True)
```

# Analysing data #
## The likelihood functions ##
Piccard offers a multitude of different likelihood functions. They are all different approximations of the same likelihood function, representing random Gaussian process. However, some are more efficient than others, and some cannot deal with specific types of signals. They are labelled mark1-mark10.

* **'mark1'** The default. It calculates the likelihood without doing any woodbury expansion. Can include red signals and DM variations. Can be combined with compression
* **'mark2'** Only white noise (efac + equad). Fastest
* **'mark3'** Woodbury expansion for red signals (noise or other) only. No DM variations. Can be combined with compression
* **'mark3fa'** Same as mark3, but uses a first-order approximation for pulsar-correlated signals. Faster than mark3 for PTAs.
* **'mark4'** Woodbury expansion for averaged TOAs (as we have for NANOGrav). Can be combined with compression. Can include red noise and DM variations. Must include a jitter noise component. Can be combined with compression (but probably not useful).
* **'mark4ln'** Same as mark4, but can include floating frequency lines. Can be combined with compression
* **'mark6'** Same as mark3, but can include DM variations. Can be combined with compression
* **'mark6fa'** Same as mark6, but uses a first-order approximation for pulsar-correlated signals.
* **'mark7'** (Unsupported) Same as mark3, but can be used in RJMCMC methods for mode-selection.
* **'mark8'** (Unsupported) Same as mark6, but can be used in RJMCMC methods for mode-selection.
* **'mark9'** (Unsupported) Same as mark3, but can include floating frequency lines. Can be combined with compression
* **'mark10'** (Unsupported) Same as mark6, but can include floating frequency lines. Can be combined with compression

Besides these choices of likelihood functions, one can set the 'compression' parameter. This uses an updated version of the ABC-compression method of [van Haasteren (2013)](http://arxiv.org/abs/1210.0584). The compression can be done either on the modelled frequencies (of both red signals and DM variations), or on the averaged TOAs. In contrast to the original ABC-compression method, the implementation of Piccard will also evaluate the complement of the likelihood function so no data is thrown away. The results are virtually identical to those of the Woodbury expansion methods, except that it is slightly faster. This method is therefore preferred, and the current default

```python
# Create a model with efac, red noise, and GWB, for dataset mdc1-open1
likob = pic.ptaLikelihood('mdc1-open1.h5')
mark1model = likob.makeModelDict(varyEfac=True, nfreqs=10, \
                                incRedNoise=True, noiseModel='powerlaw', \
                                incGWB=True, gwbModel='powerlaw', \
                                compression='frequencies', evalCompressionComplement=True, \
                                likfunc='mark1')
mark3model = likob.makeModelDict(varyEfac=True, nfreqs=10, \
                                incRedNoise=True, noiseModel='powerlaw', \
                                incGWB=True, gwbModel='powerlaw', \
                                compression=None, evalCompressionComplement=False, \
                                likfunc='mark3')
likob.initModel(mark1model)
likob.writeModelToFile('mdc1-open1.json')
pars = likob.pstart.copy()
```

```python
# Both likelihoods give approximately the same answer
likob.initModel(mark1model)
print "Mark1:", likob.logposterior(pars)
likob.initModel(mark3model)
print "Mark3:", likob.logposterior(pars)

# Note however, that for mark1, the complement of the likelihood function is evaluated
# likob.initModel(mark1model)
# print "Mark1:", likob.loglikelihood(pars), "=", likob.mark1loglikelihood(pars), "+", likob.comploglikelihood(pars)
# In the original ABC-compression paper, only 'mark1loglikelihood(pars) was evaluated.
# The implementation in Piccard is such that the user should not worry about it.
```

## Finding more optimal compression levels ##
The standard compression will compress down to the number of frequencies that we have included in our model. But that number is given in the JSON file, which ultimately is made with 'makeModelDict'. How can we decide, based on the data, how many frequencies we need to include in our model? This issue has been discussed in the literature. See for instance:

* [van Haasteren (2013)](http://arxiv.org/abs/1210.0584)
* [Lentati et al. (2013)](http://arxiv.org/abs/1210.3578)

In Piccard we have a simplified approach (for now). We evaluate the 99% fidelity level to recover a signal of a certain spectral shape and amplitude, and we use that. The spectral shape is a power-law with a certain slope, and the noise is assumed (conservatively) to consist only of the radiometer noise without applied efac. This is how it is done. (TODO: make this into a function...)

```python
# We assume we have written mdc1-open1.json to file above
likob = pic.ptaLikelihood('mdc1-open1.h5', 'mdc1-open1.json')
mark1dict = likob.getModelDict()

# Determine the number of frequencies using 'numfreqsFromSpectrum'
numfreqs = np.zeros(len(likob.ptapsrs), dtype=np.int)
for ii, psr in enumerate(likob.ptapsrs):
    numfreqs[ii] = psr.numfreqsFromSpectrum(noiseAmp=5e-14, noiseSi=4.33, threshold=0.99)

# Set the number of frequencies in the model dictionary, and re-initialise the likelihood object
mark1dict['numNoiseFreqs'] = numfreqs
likob.initModel(mark1dict)
likob.writeModelToFile('mdc1-open1.json')
```

```python
# The number of frequencies per pulsar now is (only half of the mode-frequencies, since we have two modes per frequency):
print [int(len(likob.ptapsrs[ii].Ffreqs)/2) for ii in range(len(likob.ptapsrs))]
```
